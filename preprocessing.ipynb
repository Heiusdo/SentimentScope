```python
# List of DataFrames
dfs = [df1_copy, df2_copy, df3_copy, df4_copy]

# Initialize empty lists
all_text = []
all_sentiment = []

# Loop through each DataFrame and append data to the lists
for df in dfs:
    all_text.extend(df['Text'].tolist())
    all_sentiment.extend(df['Sentiment'].tolist())

# Create a new DataFrame from the combined lists
merged_df = pd.DataFrame({'Text': all_text, 'Sentiment': all_sentiment})

# Display the merged DataFrame
print(merged_df)
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Download required NLTK data once
nltk.download('stopwords', quiet=True)
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # Convert to string if not already
    text = str(text) if not isinstance(text, str) else text
    
    # Remove non-English (non-ASCII) characters, keep basic letters, numbers, spaces, and apostrophes
    text = re.sub(r'[^\x00-\x7F\']', '', text)
    
    # Clean URLs, mentions, hashtags, and unwanted punctuation, but keep apostrophes
    text = re.sub(r'http\S+|@\w+|#[^\w\s]|[^\w\s\']', '', text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Split into words and filter stopwords
    words = [word for word in text.split() if word and word not in stop_words]
    
    # Join words with spaces
    return ' '.join(words)

# Apply to DataFrame
merged_df['Text'] = merged_df['Text'].apply(clean_text)
```
